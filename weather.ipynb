```python
url = 'https://raw.githubusercontent.com/CS109Hubway/classp/master/data/weather.csv'
weather = pd.read_csv(StringIO(requests.get(url).content), index_col=0)
weather.reset_index(inplace=True)

weather.drop('timestamp', axis=1, inplace=True)
weather['datetime'] = pd.to_datetime(weather.datetime)
weather['date'] = weather.datetime.map(lambda t: t.date())
weather['hour'] = weather.datetime.map(lambda t: t.hour)
weather.head()
```

```python
url = 'https://raw.githubusercontent.com/CS109Hubway/classp/master/data/tripsthrough2012.csv'
trips = pd.read_csv(StringIO(requests.get(url).content), index_col=0)
trips.start_date = pd.to_datetime(trips.start_date)
trips.end_date = pd.to_datetime(trips.end_date)
trips['year'] = trips.start_date.map(lambda t: t.year)
trips= trips.loc[trips.year == 2012]
trips['day_date'] = trips.start_date.map(lambda t: t.date())
trips['weekday'] = trips.day_date.map(lambda t: t.weekday())
trips['hour'] = trips.start_date.map(lambda t: t.hour)
trips.head()
```

```python
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (12, 8)})
np.random.seed(9221999)
```

```python
trip_g = trips.groupby(['day_date', 'hour'])
trip_hour = pd.DataFrame([trip_g['hubway_id'].count(), trip_g['weekday'].median()]).transpose()

trip_weather = pd.merge(trip_hour, weather, how='left', left_index=True, right_on=['date', 'hour'])
trip_weather.reset_index(inplace=True)
trip_weather['week']=trip_weather.weekday < 5
trip_weather['precip_binary']=trip_weather.precip > 0

hourly_avgs = trip_weather.groupby(['week','precip_binary','hour'])['hubway_id'].mean()
# hourly_avgs.loc[hourly_avgs['week'] == True & hourly_avgs['precip_binary'] == False]
plt.figure(figsize=(12,8))
no_rain = hourly_avgs.loc[True,False].plot(kind='bar', color="#6495ED", alpha = 0.5)
rain = hourly_avgs.loc[True,True].plot(kind='bar', color="#F08080", alpha = 0.5)
plt.legend((no_rain,rain),('No Rain', 'Rain', 'Label 2'), bbox_to_anchor=(1, 1), loc=2, fontsize = 15)

plt.figure(figsize=(12,8))
hourly_avgs.loc[False,False].plot(kind='bar', color="#6495ED", alpha = 0.5)
hourly_avgs.loc[False,True].plot(kind='bar', color="#F08080", alpha = 0.5)



# plt.figure(figsize=(12,8))
# plt.scatter(trip_weather.temp,trip_weather.hubway_id, alpha=0.5, s = 30)
# plt.xlim(0,100)
# plt.ylim(0,500)

# plt.figure(figsize=(12,8))
# plt.scatter(trip_weather.precip,trip_weather.hubway_id, alpha=0.5, s = 30)
```
